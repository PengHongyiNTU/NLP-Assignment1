```python
import re
def load_log(dir):
    train_losses = []
    valid_losses = []
    regex_loss = r"\|\sloss.*?([\d\.]+)"
    regex_valid_loss = r"valid loss.*?([\d\.]+)"
    with open(dir, 'r') as file:
        for line in file:
            train_loss = re.findall(regex_loss, line)
            if train_loss:
                train_losses.append(float(train_loss[0]))
            else:
                valid_loss = re.findall(regex_valid_loss, line)
                if valid_loss:
                    valid_losses.append(float(valid_loss[0]))
    return train_losses, valid_losses
```

```python
import matplotlib.pyplot as plt
fnn_log_dir = 'train-fnn-log.txt'
train_losses, valid_losses = load_log(fnn_log_dir)
```

```python
import numpy as np
fig, (ax1, ax2) = plt.subplots(2, figsize=(12, 10))
ax1.plot(np.arange(1, len(train_losses)+1)*100, train_losses)
ax1.set_xlabel('# of Batches')
ax1.set_ylabel('Train Loss')
ax2.plot(np.arange(1, len(valid_losses)+1), valid_losses, '-o')
ax2.set_xlabel('# of Epochs')
ax2.set_ylabel('Valid Loss')
fig.suptitle('FNN')
plt.show()
```

```python
log_dir = 'train-fnn-dropout-log.txt'
train_losses, valid_losses = load_log(log_dir)
fig, (ax1, ax2) = plt.subplots(2, figsize=(12, 10))
ax1.plot(np.arange(1, len(train_losses)+1)*100, train_losses)
ax1.set_xlabel('# of Batches')
ax1.set_ylabel('Train Loss')
ax2.plot(np.arange(1, len(valid_losses)+1), valid_losses, '-o')
ax2.set_xlabel('# of Epochs')
ax2.set_ylabel('Valid Loss')
fig.suptitle('FNN with Dropout Layer')
plt.show()
```
